# Key `Uzawa` mechanism

Plots the long-run capital market in the $(K,q)$-plane for the neoclassical- and intangible era respectively. 

The notebook calls `QP_diagram` form the python library, which computes the endogenous capital damand and -supply elasticities as per Proposition 1:
$$
    \varepsilon^S \equiv \begin{bmatrix}
            \beta_K & \beta_{L} & 0 
        \end{bmatrix} \cdot A(0)^{-1}b, \\
    \varepsilon^D \equiv  \frac{1}{\alpha_L} -\begin{bmatrix}
             \frac{s_K}{1-s_K} & - \frac{s_{L}}{1-s_{L}} & 0
        \end{bmatrix} \cdot A(0)^{-1}b,
$$
then constructs curves around the long-run steady state.
